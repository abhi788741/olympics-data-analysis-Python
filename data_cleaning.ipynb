```python
df.dropna(inplace=True)
```

```python
df=df.drop_duplicates()
```

```python
df.drop(columns="Country_Code",axis=1,inplace=True)
```

```python
df.drop(columns="Event_gender",axis=1,inplace=True)
```

```python
df
```

```python
df=df.reset_index()
```

```python
df.drop(columns="index",inplace=True)
```

```python
df
```

```python
df["Year"]=df["Year"].astype(int)
```

```python
df["Gender"].replace({"Women":"Female","Men":"Male"},inplace=True)
```

```python
df["Athlete"]=df["Athlete"].str.strip()
df["Athlete"]=df["Athlete"].str.replace(",","")
df["Athlete"]=df["Athlete"].str.capitalize()
```

```python
df["Event"]=df["Event"].str.strip()
df["Event"]=df["Event"].str.capitalize()
df["Discipline"]=df["Discipline"].str.strip()
```

```python
df
```

```python
df[df["Country"].str.contains(",")]
```

```python
df["Country"]=df["Country"].replace({"Korea, North":"North Korea","Korea, South":"South Korea"})
```
